# 13. 함수 적용

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

```python
# 키 뒤에 cm을 붙이는 함수
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
# 첫 글자를 대문자로 나머지는 소문자로 변환
df['SW특기'].str.capitalize()
```

```python
# 첫 글자를 대문자로 만들어 주는 함수 정의
def capitalize(lang):
    if pd.notnull(lang): # NaN이 아닌지
        return lang.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
school = df.groupby('학교')
school['학년'].value_counts()  # 학교로 그룹화를 한 뒤에 학년별 학생 수를 가져옴
```

```python
school = df.groupby('학교')
school['학년'].value_counts().loc['북산고']  # 학교로 그룹화를 한 뒤에 북산고 학생의 학년별 학생 수를 가져옴
```

```python
school['학년'].value_counts().loc['능남고']
```

```python
school['학년'].value_counts(normalize = True).loc['북산고'] # 학생들의 수 데이터를 퍼센트로 비교하여 가져옴 (북산고의)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

grp = df.groupby('학교')
value = [grp.size()['북산고'], grp.size()['능남고']]
labels = ['북산고','능남고']
colors = ['pink', 'violet']

wedgeprops = {'width' : 0.5, 'edgecolor' : 'w', 'linewidth' : 2}
plt.pie(value, labels = labels, wedgeprops = wedgeprops, colors = colors, autopct = '%.1f%%',pctdistance = 0.7
       , startangle = 90, counterclock = False )
plt.legend(loc=(1.2,0.3), title = '학교')
plt.title('소속 학교')
plt.show()
```
